```python
import requests

def obtener_ids_busqueda(termino, limite=50):
    # Definimos la URL de la API con el término de búsqueda y el límite de resultados
    url = "https://api.mercadolibre.com/sites/MLA/search"
    params = {"q": termino, "limit": limite}

    # Hacemos la solicitud GET
    response = requests.get(url, params=params)

    # Verificamos si la solicitud fue exitosa (código de respuesta 200)
    if response.status_code == 200:
        # Convertimos la respuesta a formato JSON
        data = response.json()
        # Extraemos los IDs de los ítems de los resultados de la búsqueda
        ids = [item["id"] for item in data.get("results", [])]
        return ids
    else:
        print("Error al realizar la solicitud:", response.status_code)
        return []

# Términos de búsqueda
terminos = ["Google Home TV", "Apple TV", "Amazon Fire TV"]

# Recopilamos los IDs de los ítems para cada término de búsqueda
for termino in terminos:
    ids = obtener_ids_busqueda(termino)
    print(f"IDs de la búsqueda '{termino}': {ids}")
```

```python
import csv

def escribir_resultados_a_csv(resultados, nombre_archivo):
    # Definimos los nombres de los campos que queremos incluir en el archivo CSV
    campos = ['id', 'title', 'category_id', 'price', 'currency_id', 'condition', 'permalink', 'thumbnail']

    # Abrimos el archivo CSV en modo de escritura
    with open(nombre_archivo, 'w', newline='', encoding='utf-8') as archivo_csv:
        # Creamos un escritor CSV
        escritor_csv = csv.DictWriter(archivo_csv, fieldnames=campos)

        # Escribimos los encabezados
        escritor_csv.writeheader()

        # Escribimos cada fila de datos en el archivo CSV
        for resultado in resultados:
            escritor_csv.writerow(resultado)

# Lista para almacenar los resultados
resultados = []

# Obtener y almacenar la información de los ítems de Google Home TV
for item_id in ids_google_home_tv:
    info = obtener_informacion_item(item_id)
    if info:
        resultados.append({
            'id': info.get('id', ''),
            'title': info.get('title', ''),
            'category_id': info.get('category_id', ''),
            'price': info.get('price', ''),
            'currency_id': info.get('currency_id', ''),
            'condition': info.get('condition', ''),
            'permalink': info.get('permalink', ''),
            'thumbnail': info.get('thumbnail', '')
        })

# Obtener y almacenar la información de los ítems de Apple TV
for item_id in ids_apple_tv:
    info = obtener_informacion_item(item_id)
    if info:
        resultados.append({
            'id': info.get('id', ''),
            'title': info.get('title', ''),
            'category_id': info.get('category_id', ''),
            'price': info.get('price', ''),
            'currency_id': info.get('currency_id', ''),
            'condition': info.get('condition', ''),
            'permalink': info.get('permalink', ''),
            'thumbnail': info.get('thumbnail', '')
        })

# Obtener y almacenar la información de los ítems de Amazon Fire TV
for item_id in ids_amazon_fire_tv:
    info = obtener_informacion_item(item_id)
    if info:
        resultados.append({
            'id': info.get('id', ''),
            'title': info.get('title', ''),
            'category_id': info.get('category_id', ''),
            'price': info.get('price', ''),
            'currency_id': info.get('currency_id', ''),
            'condition': info.get('condition', ''),
            'permalink': info.get('permalink', ''),
            'thumbnail': info.get('thumbnail', '')
        })

# Escribir los resultados en un archivo CSV llamado "resultados.csv"
escribir_resultados_a_csv(resultados, 'resultados.csv')
```
